```python
dates_df = pd.date_range(
    dt.strptime("01012023", "%d%m%Y"), dt.strptime("31122025", "%d%m%Y")
)

cal_df = pd.DataFrame()
cal_df["year"] = dates_df.year
cal_df["month"] = dates_df.month
cal_df["day"] = dates_df.day

cal_df["isHolyday"] = 0
for key, val in holidays().items():
    for holiday in val:
        cal_df["isHolyday"] = np.where(
            (cal_df["year"] == int(key))
            & (cal_df["month"] == holiday["month"])
            & (cal_df["day"] == holiday["day"]),
            1,
            cal_df["isHolyday"],
        )

cal_df["isWorkingDay"] = 1
for key, val in weekend().items():
    for weekends in val:
        for day in weekends["days"]:
            cal_df["isWorkingDay"] = np.where(
                (cal_df["year"] == int(key))
                & (cal_df["month"] == weekends["month"])
                & (cal_df["day"] == day),
                0,
                cal_df["isWorkingDay"],
            )

isLastWorkingDays = []
last_val = 0
for val in cal_df["isWorkingDay"].values[-1::-1]:
    isLastWorkingDays.append(1 if (last_val == 0 and val == 1) else 0)
    last_val = val

cal_df["isLastWorkingDay"] = isLastWorkingDays[-1::-1]
```
